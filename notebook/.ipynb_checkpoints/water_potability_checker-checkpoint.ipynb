# Water Potability Checker

## 1. Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
water_df = pd.read_csv('../dataset/water_potability.csv')
water_df.head()
```

## 2. Data Preparation

### 2.1 Assessing data

```python
print(f"The dataframe has {water_df.shape[0]} samples of {water_df.shape[1]} features each.")
```

```python
water_df.dtypes
```

All columns are numeric (float64 or int64). So there will not a categorical encoding.

The potability is the column that says if a water is potable or not. Logically it will have 2 values (0 and 1). Let's see that.

```python
water_df['Potability'].unique()
```

There are indeed 2 unique values in that column.

```python
cols_with_missing = [col for col in water_df.columns if water_df[col].isnull().any()]
cols_with_missing
```

The columns **ph**, **Sulfate** and **Trihalomethanes** have missing values.

Let's handle these missing values.

### 2.2. Handle missing values

```python
print("Proportion of missing values per column")
water_df.isnull().sum()
```

There are three columns of the 9 features (features are columns except Potability) that contain null values.  

```python
water_df['nnull'] = water_df.isnull().sum(axis=1)
```

```python
labels = [0, 1, 2, 3]
proportions = []

for label in labels:
    l_null = water_df.query(f'nnull == {label}')
    n_samples = l_null.shape[0]
    proportions.append(l_null.shape[0])
    
proportions = pd.Series(proportions)
proportions
```

```python
plt.bar(list(map(str, labels)), proportions, color=['blue', 'red', 'yellow'])
plt.xlabel('Number of null values')
plt.ylabel('Number of samples (lines)')
plt.title('Number of null values by number of samples');
```

It is obvious that the more there is null values on a specific line, the less correct will be the relationship between the features and the target on this line.  
With the objective to have correct data, only the lines that have less than three null values will be considered.

```python
water_df.query('nnull <= 2', inplace=True)
```

The **nnull** column should be droped because it is no longer useful.

```python
water_df.drop(['nnull'], axis=1, inplace=True)
```

Let's impute values for the remaining missing values.

```python
water_df[['ph', 'Sulfate', 'Trihalomethanes']].hist(figsize=(8,8));
```

By looking at the distributions of the variables with missing values, it appears they all follow a normal distribution. Then, **impute the mean** is a good idea. 

```python
water_df['ph'].fillna(water_df['ph'].mean(), inplace=True)
water_df['Sulfate'].fillna(water_df['Sulfate'].mean(), inplace=True)
water_df['Trihalomethanes'].fillna(water_df['Trihalomethanes'].mean(), inplace=True)
```

### 2.3. Remove duplicated values

```python
water_df.duplicated().sum()
```

```python
water_df.duplicated([col for col in water_df.columns if col != 'Potability']).sum()
```

There is no duplicated values in the dataset.

### 2.4. Handle outliers

```python
water_df.hist(figsize=(15,15));
```

All the features seem to follow a normal distribution. But the distribution of **Solids** is  **right-skewed**. Let's apply the **Interquartile Range (IQR) Proximity rule** to check if there are outliers in the column of Solids.

```python
IQR = water_df['Solids'].quantile(0.75) -  water_df['Solids'].quantile(0.25)
IQR
```

```python
lower_limit = water_df['Solids'].quantile(0.25) - (IQR*3)
lower_limit
```

```python
upper_limit = water_df['Solids'].quantile(0.75) + (IQR*3)
upper_limit
```

```python
water_df_rem = water_df.query(f'Solids >= {lower_limit} and Solids <= {upper_limit}')
len(water_df) == len(water_df_rem)
```

There are apparently no outliers in this column.

### 2.5. Feature engineering

```python
plt.figure(figsize=(9,6))
sns.heatmap(water_df.corr(), cbar=True, annot=True);
plt.title('Correlation between the variables');
```

It seems there is no correlation between the variables of the dataset. That supposes an absence of linear relationship between the different features. 
Let's try to find if there are other types of relationship between the variables by viewing their mutual information.

```python
features = water_df.drop(['Potability'], axis=1)
target = water_df['Potability']
```

```python
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(features, target)
mi_scores = pd.Series(mi_scores, name='MI Scores', index=features.columns)
mi_scores = mi_scores.sort_values(ascending=True)
mi_scores
```

By looking at the mutual information of the different columns, there are no strong relationships between the features and the target. The fact that **no mi_score is near to 0.5** is the proof.

## 3. Modeling

### 3.1. Split data

```python
target_yes = (target == 1).sum()
target_no = (target == 0).sum()
```

```python
plt.pie([target_yes, target_no], labels=['yes', 'no'], autopct='%1.1f%%', colors=['green', 'red']);
```

The proportion of 'no' values is almost twice the proportion of 'yes' values. Therefore, the splitting should be done carefully.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, train_size=0.8, stratify=target)
```

### 3.2. Training and Testing

For the training, we will use a data-centric approach that means no hyperparameter tuning should be done. We will work with three models: a **Logistic Regression** model, a **Decision Tree** Classifier and a **Random Forest Classifier**.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
models_init = {
    'LR': LogisticRegression(),
    'DTC': DecisionTreeClassifier(),
    'RFC': RandomForestClassifier()
}
```

```python
models_trained = {}

for name, model in models_init.items():
    model.fit(X_train, y_train)
    models_trained[name] = model
    y_pred = model.predict(X_test)
    score = accuracy_score(y_pred, y_test)
    print(f"{name} model score: {np.round(score*100, 2)}%")  
```

Let's use cross validation to choose the right model.

```python
from sklearn.model_selection import cross_val_score

for name, model in models_init.items():
    score_l = cross_val_score(model, features, target, cv=5)
    print(f"The accuracy for the {name} model is {np.round(np.average(score_l)*100, 2)}%")  
```

By comparing the scores with simple validation and cross validation, it appears that the **Random Forest model** performs better.

### 3.3. Model saving

```python
import pickle

filename = 'model/rf_model.pkl'
pickle.dump(models_trained['RFC'], open(filename, 'wb'))
```

## 4. Conclusion

The score of the model is not bad (about **63%**) but it proves the features describe poorly the target. As demonstred by the mutual information and correlation data, the features have not that strong relationship with the target.  
Nevertheless, some improvements can be done to the model to enhance its performance. For example: find **more data**, use a **Gradient Boosted Tree (GDT)** model, **feature scaling**, **hyperparameter tuning**, etc.
